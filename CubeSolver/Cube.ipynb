## 2. Q-Learning visualization and test

This part loads the 100 Train-Test rounds information with the learned Q-Table. First, visualization of the Train-Test history will be done, then after loading the learned Q-Table, some simple tests will be done.

```python
## load the Train-Test history
import pickle
with open('Q.pickle') as f:
    qsizes, qtable, result = pickle.load(f)
## import for plotting
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
## plot the recovery rates for permutation lengths = [1,2,3,4,5,7,9,12,16] overtime
rr = {}
lengths = [1,2,3,4,5,7,9,12,16]
for length in lengths:
    rr[length] = []
    for r in range(0,100):
        rr[length].append(1.*result[r]['rc'][length]/result[r]['tc'][length])
plt.figure(figsize=(8, 6))
for length in lengths:
    x = np.asarray([r for r in range(1,101)])
    y = np.asarray([rr[length][i] for i in range(0,100)])
    plt.plot(x,y,label=str(length))
plt.xlabel('rounds')
plt.ylabel('recovery rates')
plt.title('Q-Learning recovery rates')
plt.legend()
plt.show()
## plot the average recovery moves for permutation lengths = 5 overtime
am5 = []
for r in range(0,100):
    am5.append(result[r]['sl'][5])
plt.figure(figsize=(8,6))
plt.plot(np.asarray([r for r  in range(1,101)]),np.asarray(am5))
plt.xlabel('rounds')
plt.ylabel('average recovery moves for pl=5')
plt.title('Q-Learning average recovery moves for permutation length 5')
plt.show()
## plot the Q-Table size overtime
plt.figure(figsize=(8,6))
plt.plot(np.asarray([r for r  in range(1,101)]),np.asarray(qsizes))
plt.xlabel('rounds')
plt.ylabel('Q-Table size')
plt.title('Q-Learning Q-Table size')
plt.show()
```

It can be seen that overtime the recovery rates for different permutation length test cases go up, the average moves used to solve go down, and the Q-Table size increase more or less linearly.

```python
actions = ['F','B','L','R','U','D','f','b','l','r','u','d']
## tests
import random
## manual inspection of the Q-Learning's performance on the following test cases
for testPermutation in ['F','d','Fu','dl','LBd','DDD','FRUBL','dlbur']:
    solved = False
    moves = testPermutation
    for t in range(0, 10): # let's give the QNN 10 tries to solve each test permutation
        myCube.reset()
        myCube.do(moves)
        state = ''.join(myCube.facelets)
        if myCube.good():
            solved = True
            break
        else:
            if state not in qtable:
                action = random.choice(actions)
            else:
                action = max(qtable[state],key=qtable[state].get)
            moves = moves + action
    print 'solved: %s, permutation: %s, solveMoves: %s' % (solved, testPermutation, moves[len(testPermutation):])
```

* The Q-Learning agent manages to solve all 8 test cases
* It recovers F with f, d with D, dl with LD, which are direct reverse actions
* It recovers Fu with Dl, DDD with D, which means that the agent also finds alternative ways to recover the cube to a good state
* For more complex cases FRUBL and dlbur, the agent solves it with ways that are not instantly obvious for me, but after manually checking them with the help of [ruwix](https://ruwix.com/online-puzzle-simulators/2x2x2-pocket-cube-simulator.php), it does fully recover the positions. Note that both the recovered positions are no the original cube
* The agent only takes color information on the 6 faces as input, it's really interesting to see that it can recover these test cases

## 3. modified Q-Learning visualization and test

This part loads the 3 Train-Test rounds information with the learned QNN. First, visualization of the Train-Test history will be done, then after loading the learned QNN, some simple tests will be done.

```python
## load the Train-Test history
import pickle
with open('QNN.pickle') as f:
    result = pickle.load(f)
## load the QNN model
from keras.models import load_model
import numpy as np
model = load_model('QNNmodel')
## import for plotting
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
## since the training time for QNN is too long, only 3 Train-Test rounds were recorded
## therefore here we only plot the overall recovery rate
## 20 permutation lengths, 100 test cubes each permutation length, 2000 test cubes each round

## calculate the overall recovery rates
rr = [0,0,0]
for i in range(0,3):
    testCounts = 0
    recoveryCounts = 0
    for pl in range(1,21):
        testCounts = testCounts + result[i]['tc'][pl]
        recoveryCounts = recoveryCounts + result[i]['rc'][pl]
    rr[i] = 1. * recoveryCounts / testCounts
## plot the rates
x = np.asarray([i for i in range(1,4)])
y = np.asarray([rr[i] for i in range(0,3)])
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('rounds')
plt.ylabel('overall recovery rate')
plt.title('QNN overall recovery rate')
plt.show()
```

```python
actions = ['F','B','L','R','U','D','f','b','l','r','u','d']
## help function that translate string state to np vector for model to predict on
def state2x(state):
    x = np.zeros([1,144])
    colorDict = {'g':[0,0,0,0,0,1],'b':[0,0,0,0,1,0],'o':[0,0,0,1,0,0],\
                 'r':[0,0,1,0,0,0],'w':[0,1,0,0,0,0],'y':[1,0,0,0,0,0]}
    for i in range(0,24):
        x[0,6*i:6*i+6] = colorDict[state[i]]
    return x
## tests
## manual inspection of the QNN's performance on the following test cases
for testPermutation in ['F','d','Fu','dl','LBd','DDD','FRUBL','dlbur']:
    solved = False
    moves = testPermutation
    for t in range(0, 10): # let's give the QNN 10 tries to solve each test permutation
        myCube.reset()
        myCube.do(moves)
        state = ''.join(myCube.facelets)
        if myCube.good():
            solved = True
            break
        else:
            action = actions[np.argmax(model.predict(state2x(state)))]
            moves = moves + action
    print 'solved: %s, permutation: %s, solveMoves: %s' % (solved, testPermutation, moves[len(testPermutation):])
```

* Out of the 8 tests the QNN managed to solve 7 of them.
* For the first 6 tests, it's obvious that the agent mainly solves them by reversing the permutation process, and when the cube complexity is reduced to a simple state, it solves it by one face rotation that's not necessarily reversing the permutation (it solves F with b instead of f).
* Note that the agent only takes facelet color as input, it has no knowledge of what the permutation was.
* For the failed case, the agent doesn't seem to recognize the position, so it turns the UP face again and again hoping to result in a position that it knows how to solve, but unfortunately it fails to find one
* For the last test case, the agent successfully solves the cube, what's interesting is that **it doesn't solve by reversing the permutation process**, which should be "RUBLD", instead, it chooses to solve with "UBLUF". This means that the QNN agent not only learns to solve by "memorizing", but also discovered some alternative paths. Try following the permutation and the solveMoves in [ruwix](https://ruwix.com/online-puzzle-simulators/2x2x2-pocket-cube-simulator.php), you will find that the agent solves this cube to a good position that's rotated from the original cube.
